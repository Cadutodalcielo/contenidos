Importa las librerias necesarias

```
import altair as alt
import pandas as pd
```

Carga los datos desde la url dada, codigo obtenido de la ayudantia 

```
url = "https://raw.githubusercontent.com/Exploratorio-DCC-PUC/Syllabus/master/Tareas/TG1/datos.csv"

data = pd.read_csv(url)
data = data.drop(columns=["Unnamed: 0", "ID", "Type"])
# data
```

# Parte 1

## Pregunta 1 (IMDB)

### IMDB vs Todas las películas

```
# Pandas

# borra los valores nulos
imdb_vs_tp = data[~(data["IMDb"].isnull())]
imdb_vs_tp["IMDb"] = imdb_vs_tp["IMDb"].astype(int)
# selecciona solo las columnas IMDb y Title
imdb_vs_tp = imdb_vs_tp[["IMDb", "Title"]]
# redondea los numeros a 0 decimales
imdb_vs_tp = imdb_vs_tp.round(0)
# cambia todos los valores de la columana title a 1
imdb_vs_tp["Title"] = 1
# agrupa por los criterios de IMDb y luego calcula la suma
imdb_vs_tp = imdb_vs_tp.groupby(by = "IMDb").sum()

# No podia ver el gráfico bien, estuve buscando 4 horas el problema, al final 
# no funcionaba porque al hacer groupby "IMDb" lo cuenta como indice y no como
# columna por lo que cuando le indicamos esa columna a altair no la encuentra,
# la solución es resetear el indice para que IBDb vuelva a ser columna.
# 
# si te lo preguntabas la respuesta es si, estuve cuatro horas para solucionar 
# el promebla con  literalmente 1 linea  >:C . Por lo menos aprendi harto de 
# altair xd.
imdb_vs_tp.reset_index(level=0, inplace=True)


# Altair 

alt.Chart(imdb_vs_tp).mark_bar().encode(alt.X('IMDb:Q', title = "IMDB"), alt.Y('Title:Q', title = "Todas las películas")).properties(title = "IMDB vs Todas las películas")
```

### IMDB vs Netflix

## Pregunta 1 (Rotten tomatoes)

### Rotten Tomatoes vs Todas las películas

```
# Pandas


rt_vs_tp = data[~(data["Rotten Tomatoes"].isnull())]
rt_vs_tp["Rotten Tomatoes"] = rt_vs_tp["Rotten Tomatoes"].str.replace("%", "")
rt_vs_tp["Rotten Tomatoes"] = rt_vs_tp["Rotten Tomatoes"].astype(int)
rt_vs_tp = rt_vs_tp[["Rotten Tomatoes", "Title"]]

rt_vs_tp["Title"] = 1

rt_vs_tp = rt_vs_tp.groupby(by = "Rotten Tomatoes").sum()

rt_vs_tp.reset_index(level=0, inplace=True)

rt_vs_tp = rt_vs_tp.sort_values("Title", ascending=False)

# Altair 

alt.Chart(rt_vs_tp).mark_bar().encode(alt.X('Rotten Tomatoes:N', title = "Rotten Tomatoes"), alt.Y('Title:Q', title = "Todas las películas")).properties(title = "Rotten Tomatoes vs Todas las películas")
```

## Pregunta 2

```
rt_vs_imdb = data[~(data["IMDb"].isnull()) & ~(data["Rotten Tomatoes"].isnull()) & ~(data["Language"].isnull()) & ~(data["Runtime"].isnull())]
rt_vs_imdb = rt_vs_imdb[["IMDb", "Rotten Tomatoes", "Language", "Runtime"]]
rt_vs_imdb["Rotten Tomatoes"] = rt_vs_imdb["Rotten Tomatoes"].str.replace('%', '', regex=True)
rt_vs_imdb["Rotten Tomatoes"] = rt_vs_imdb["Rotten Tomatoes"].astype(int)
rt_vs_imdb["Runtime"] = rt_vs_imdb["Runtime"].astype(int)
rt_vs_imdb = rt_vs_imdb.sort_values('Rotten Tomatoes', ascending=False)
rt_vs_imdb["Language"] = rt_vs_imdb["Language"].str.split(",").str[0]


# rt_vs_imdb
# altair
alt.data_transformers.disable_max_rows()
alt.Chart(rt_vs_imdb).mark_circle().encode(x = alt.X('IMDb:Q', title = "IMDB"), y = 'Rotten Tomatoes:Q', color = alt.Color('Language:N', sort = ["English", "Spanish", "French"]), size = alt.Size('Runtime:N'))\
.configure_legend(symbolLimit = 10).properties(title = "IMDB vs Rotten Tomatoes")
```
